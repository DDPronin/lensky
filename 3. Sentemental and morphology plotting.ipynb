**Построение кривых эмоциональной тональности, частотности частей речи и облака слов**

Импорт Lensky

```python
from Lensky.lensky import *
```

Загрузка корпуса по адресу

```python
corp = Corpus('data/corpus_ru')
```

```python
corp['Лермонтов'].keys()
```

```python
text = corp['Лермонтов']['Лермонтов.Герой_нашего_времени.1840']
```

Построение кривой эмоциональной тональности

```python
sentiment_plot(text, plot_type='pos-neg', window=200, grid=True, zero_level=False, title='Эмоциональная тональность "Героя нашего времени"', figsize=(8, 5), dpi=150)
```

Построение частотности для глагола

```python
morph_plot(text, plot_type=['VERB'], window=500, title='Глаголы в произведении "Герой нашего времени"', grid=True)
```

Загрузим корпус с удаленными стоп-словами

```python
corp_norm = Corpus('data/corpus_ru_normalized')
```

```python
text = corp_norm['Лермонтов']['Лермонтов.Герой_нашего_времени.1840']
```

Построение облака слов

```python
plot_wordcloud(text, stopwords=LENSKY_STOP_WORDS)
```

Удалим оставшиеся стоп-слова, которые мешают восприятию

```python
ignore = ['сказать', 'который', 'мой', 'свой', 'который', 
          'говорить', 'отвечать', 'рука', 'хотеть', 'отвечать', 
          'мочь']
```

```python
plot_wordcloud(text, stopwords=LENSKY_STOP_WORDS+ignore)
```
