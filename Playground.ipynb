```python
def SIF_Embedding(X,p,a=1e-3):
    # X is the concatenated word embeddings (n,m) 
    # p is the probability of each word     (n,1)
            
    U,_,_ = np.linalg.svd(np.matmul(X.transpose(),X))
    U = U[:,0]
        
    mV = np.matmul( ( a / (a + p) ) , X)
    mV /= X.shape[0]
    mV -= U * np.matmul(U.transpose(), mV)
    
    return mV
```
