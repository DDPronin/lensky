```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import tqdm
```

```python
df = pd.read_excel('data//RuLitStat.xlsx')
df
```

```python
syn = df.drop(np.append(df.columns[:4], 'cent'), axis=1).to_numpy()
syn
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
syn = scaler.fit_transform(syn)
dfd = pd.DataFrame(data=syn, columns=df.columns[4:].drop('cent'))
dfd = pd.concat([df[np.append(df.columns[:4], 'cent')], dfd], axis=1)
dfd
```

```python
%%time
DATA = sorted(list(df['filename']))
AUTHORS = sorted(list(df['author']))
corpus = {}
for author in tqdm.tqdm(AUTHORS):
    author_books = []
    aut = {}
    for i in range(len(DATA)):
        if author == DATA[i].split('.')[0]:
            aut['.'.join(DATA[i].split('.')[:-1])] = dfd[dfd['filename']==DATA[i]].drop(np.append(dfd.columns[:4], 'cent'), axis=1).to_numpy()[0]
    corpus[author] = aut
```

```python
def model_fit(corpus, model):
    list_ = []
    print(f'Model fitting...')
    for author in tqdm.tqdm(list(corpus.keys())):
        for book in list(corpus[author].keys()):
            list_.append(corpus[author][book])
    model.fit(list_)
    print('Done!')
```

```python
import umap

mapper = umap.UMAP()
model_fit(corpus, mapper)
```

```python
from matplotlib.pyplot import figure
def build_model_plot(corpus, model, oper=None):
    figure(figsize=(10, 10), dpi=250)
    print('Corpus processing...')
    vecs = []
    for author in tqdm.tqdm(list(corpus.keys())):
        for book in list(corpus[author].keys()):
            vecs.append(corpus[author][book])
    z_model = model.transform(vecs)
    n = 0
    c18, c19, c20, c21 = [], [], [], []
    for author in tqdm.tqdm(list(corpus.keys())):
        for book in list(corpus[author].keys()):
            if int(book.split('.')[-1]) < 1801:
                c18.append(n)
            if int(book.split('.')[-1]) < 1901 and int(book.split('.')[-1]) >= 1801:
                c19.append(n)
            if int(book.split('.')[-1]) < 2001 and int(book.split('.')[-1]) >= 1901:
                c20.append(n)
            if int(book.split('.')[-1]) >= 2001:
                c21.append(n)
            n+=1
    print('Done!')
    plt.scatter(z_model[c18, 0], z_model[c18, 1], s=7, color='C0', alpha=0.6, label='XVIII')
    plt.scatter(z_model[c19, 0], z_model[c19, 1], s=7, color='C1', alpha=0.6, label='XIX')
    plt.scatter(z_model[c20, 0], z_model[c20, 1], s=7, color='C2', alpha=0.6, label='XX')
    plt.scatter(z_model[c21, 0], z_model[c21, 1], s=7, color='C3', alpha=0.6, label='XXI')
    plt.title(f'Век написания: {oper} отображение')
    plt.legend(title='Век написания', loc='upper center', bbox_to_anchor=(0.5, -0.05))
    plt.xticks([])
    plt.yticks([])
```

```python
build_model_plot(corpus, mapper, oper='UMAP')
```

```python
from matplotlib.pyplot import figure
def build_books(corpus, target, model, oper=None):
    figure(figsize=(10, 10), dpi=250)
    print('Corpus processing...')
    vecs = []
    for author in tqdm.tqdm(list(corpus.keys())):
        for book in list(corpus[author].keys()):
            if author != target:
                vecs.append(corpus[author][book])
    z_model = model.transform(vecs)
    plt.scatter(z_model[:,0], z_model[:, 1], color='gray', s=7, alpha=0.3)
    print('Done!')
    print('Author books processing...')
    author_vecs = []
    for book in list(corpus[target].keys()):
        author_vecs.append(corpus[target][book])
    z_model = model.transform(author_vecs)
    for n, book in tqdm.tqdm(enumerate(corpus[target])):
        plt.scatter(z_model[n, 0], z_model[n, 1], label=' '.join((book.split('.')[1]).split('_')), s=40, marker=f'${n+1}$')
    print('Done!')
    plt.title(f'{target}: {oper} отображение')
    plt.legend(title='Произведение', loc='upper center', bbox_to_anchor=(0.5, -0.05))
    plt.xticks([])
    plt.yticks([])
```

```python
build_books(corpus, 'Пелевин', mapper, 'UMAP')
```
