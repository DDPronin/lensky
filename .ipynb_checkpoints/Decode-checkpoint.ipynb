```python
import os
```

```python
DATA_PATH = 'D:\\DDPronin\\PyCharm\\RU Classic Literature graphs\\data\\corpus_ru'
DATA = os.listdir(DATA_PATH)
DATA
```

```python
for file in DATA:
    file_name = DATA_PATH + '\\' + file
    with open(file_name, encoding='utf-8-sig', errors='ignore') as f:
        text = f.read().rstrip()
    with open(file_name, 'w', encoding='utf-8') as f:
        f.write(text)
```
