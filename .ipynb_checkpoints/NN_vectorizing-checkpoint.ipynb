```python
DATA_PATH = 'D:\\DDPronin\\PyCharm\\RU Classic Literature graphs\\data\\corpus_ru_tokenized'
DATA = os.listdir(DATA_PATH)
DATA
```

```python
AUTHORS = list(set([element.replace('_','.').split('.')[0].lower() for element in DATA]))
```

```python
#Функция для считывания книг в словарь
def to_dict(DATA_PATH, BOOKS):
    out_dict = {}
    for name in BOOKS:
        with open(f'{DATA_PATH}\\{name}', encoding="utf-8") as f:
            out_dict[name[:-4]] = ' '.join(f.read().split())
    return out_dict
```

```python
%%time
corpus = {}
data_low = [data.lower() for data in DATA]
for author in tqdm.tqdm(AUTHORS):
    author_books = []
    for i in range(len(data_low)):
        if author in data_low[i]:
            author_books.append(DATA[i])
    corpus[author] = to_dict(DATA_PATH, author_books)
```

```python
# {key: [el1, el2...], ...} -> {key: [vec1, vec2...], ...}
def vectorize_navec(corpus, navec):
    keys = list(corpus.keys())
    books = list(itertools.chain.from_iterable(corpus.values())) # {key: [el1, el2...], ...} -> [el1, el2, el3, el4...]
    #Векторизация
    out = {}
    for key in tqdm.tqdm(keys):
        out[key] = []
        for element in corpus[key]:
            text_vec = 0
            len_ = 0
            element = element.split()
            for word in element:
                if word in navec:
                    len_ += 1
                    text_vec += navec[word]
            out[key].append(text_vec/len_)
    return out
```

```python
sent18 = []
sent19 = []
sent20 = []
sent21 = []
keys = list(corpus.keys())
for key in keys:
    for book in corpus[key].keys():
        year = book.split('.')[-1]
        if year[-1] in [')', ']']:
            year = year[:-1]
        year = int(year)
        if year < 1801:
            sent18.append(book)
        if 1800 < year and year < 1901:
            sent19.append(book)
        if 1900 < year and year < 2001:
            sent20.append(book)
        if year >= 2001:
            sent21.append(book)
```

```python
corpus_sent_book = {18 : [], 19 : [], 20 : [], 21 : []}
for key in keys:
     for name in corpus[key].keys():
            if name in sent18:
                corpus_sent_book[18].append(corpus[key][name])
            if name in sent19:
                corpus_sent_book[19].append(corpus[key][name])
            if name in sent20:
                corpus_sent_book[20].append(corpus[key][name])
            if name in sent21:
                corpus_sent_book[21].append(corpus[key][name])
```

```python
%%time
corpus_sent_book_vecs = vectorize_navec(corpus_sent_book, navec)
```

```python
import umap

mapper = umap.UMAP()
```

```python
def fit_model(corpus, model):
    books = list(itertools.chain.from_iterable(corpus.values()))
    model.fit(books)
```

```python
fit_model(corpus_sent_book_vecs, mapper)
```

```python
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure

#Функция для отрисовки графиков
# {key: [vec1, vec2...], ...} -> plot
def build_model_plot(corpus, model, colors, point_size = 10, mass_center = False, 
                     legend_title='legend_title', plot_title='plot_title'):
    figure(figsize=(10, 10), dpi=250)
    keys = list(corpus.keys())
    for i, key in enumerate(keys):
        z_model = model.transform(corpus[key])
        plt.scatter(z_model[:,0], z_model[:,1], label=key, s=point_size, c=colors[i], linewidth=0.5, edgecolor='k')
        if mass_center:
            x = np.mean(z_model[:,0])
            y = np.mean(z_model[:,1])
            plt.scatter(x, y, label=key, s=point_size, c=colors[i], marker="X")
    plt.legend(title=legend_title)
    plt.title(plot_title)
    plt.xticks([])
    plt.yticks([])
```

```python
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure

#Функция для отрисовки графиков
# {key: [vec1, vec2...], ...} -> plot
def build_books_plot(corpus, author, books, model, colors, point_size = 10, mass_center = False, 
                     legend_title='legend_title', plot_title='plot_title'):
    figure(figsize=(10, 10), dpi=250)
    keys = list(corpus.keys())
    corpus_list = corpus.values()
    temp = []
    for i, key in enumerate(keys):
        if key != author:
            temp += corpus[key]
    z_model = model.transform(temp)
    plt.scatter(z_model[:,0], z_model[:,1], s=point_size, alpha=0.3, 
                c='gray', linewidth=0.5, edgecolor='k')
    z_model = model.transform(corpus[author])
    for i in range(len(books)):
        plt.scatter(z_model[i, 0], z_model[i, 1], label=books[i], s=40, marker=f'${i}$')
    plt.legend(title=legend_title, loc='upper center', bbox_to_anchor=(0.5, -0.05))
    plt.title(plot_title)
    plt.xticks([])
    plt.yticks([])
```

```python
from matplotlib.pyplot import figure
def build_books(vec_lists, target, books, skip_id, model, oper=None):
    figure(figsize=(10, 10), dpi=250)
    list_ = []
    for vec_list in vec_lists:
        list_ += vec_list
    model.fit(list_)
    for i, vec_list in enumerate(vec_lists):
        if i != skip_id:
            z_model = model.transform(vec_list)
            plt.scatter(z_model[:,0], z_model[:, 1], color='gray', s=7, alpha=0.3)
    i = 0
    c = 0
    for vec_list in target:
        n = 0
        for vec in vec_list:
            z_model = model.transform([vec])
            plt.scatter(z_model[:, 0], z_model[:, 1], label=books[i], s=40, marker=f'${n}$', color=f'C{c}')
            i += 1
            n += 1
        c += 1
    plt.legend(title='Novel')
    plt.title(f'Books plot {oper}')
    plt.xticks([])
    plt.yticks([])
```
