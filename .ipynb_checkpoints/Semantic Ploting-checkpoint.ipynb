```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import tqdm

import os

import re

import nltk
from nltk import word_tokenize

import codecs

import pymorphy2

from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
DATA_PATH = 'D:\\DDPronin\\PyCharm\\RU Classic Literature graphs\\data\\corpus_ru_normalized_cleared'
DATA = os.listdir(DATA_PATH)
DATA
```

```python
AUTHORS = sorted(list(set([element.replace('_','.').split('.')[0] for element in DATA])))
print('Авторов: ', len(AUTHORS))
print('Книг: ', len(DATA))
```

```python
#Функция для считывания книг в словарь
def to_dict(DATA_PATH, BOOKS):
    out_dict = {}
    for name in BOOKS:
        with open(f'{DATA_PATH}\\{name}', encoding="utf-8") as f:
            out_dict[name[:-4]] = ' '.join(f.read().split())
    return out_dict
```

```python
%%time
corpus = {}
for author in tqdm.tqdm(AUTHORS):
    author_books = []
    for i in range(len(DATA)):
        if author == DATA[i].split('.')[0]:
            author_books.append(DATA[i])
    corpus[author] = to_dict(DATA_PATH, author_books)
```

```python
def dict_apply(func, corpus):
    keys = list(corpus.keys())
    out = []
    for author in keys:
        for book in corpus[author].keys():
            out.append(func(corpus[author][book]))
    return out
```

```python
def notpop_out(book):
    print('Processing...')
    for author in tqdm.tqdm(list(corpus.keys())):
        for book in list(corpus[author].keys()):
            text = corpus[author][book]
            lim = 0.001
            pop_series = pd.Series(text.split()).value_counts(normalize=True).reset_index().to_numpy()
            pop_set = set(pop_series[pop_series[:,1]>lim][:,0])
            out = []
            for word in text.split():
                if word in pop_set:
                    out.append(word)
            corpus[author][book] = ' '.join(out)
```

```python
%%time
#оставляем в произведениях только слова, доля которых превосходит 1/1000
notpop_out(corpus)
```

```python
def vectorize(corpus, vectorizer):
    vec_corp = []
    out_dicts = []
    #Создается корпус всех произведений 
    print("Vectorizer fitting...")
    for author in tqdm.tqdm(list(corpus.keys())):
        for book in list(corpus[author].keys()):
            vec_corp.append(corpus[author][book])
    vectorizer = vectorizer.fit(vec_corp)
    print('Done!\n')
    #Векторизация
    print("Vectorizing...")
    dict_vec_list = []
    for author in tqdm.tqdm(list(corpus.keys())):
        for book in list(corpus[author].keys()):
            corpus[author][book] = vectorizer.transform([corpus[author][book]]).toarray()[0]
    print("Done!")
```

```python
%%time
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()
vectorize(corpus, tfidf)
```

```python
def model_fit(corpus, model):
    list_ = []
    print(f'Model fitting...')
    for author in tqdm.tqdm(list(corpus.keys())):
        for book in list(corpus[author].keys()):
            list_.append(corpus[author][book])
    model.fit(list_)
    print('Done!')
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
model_fit(corpus, pca)
```

```python
import umap

mapper = umap.UMAP()
model_fit(corpus, mapper)
```

```python
from matplotlib.pyplot import figure
def build_model_plot(corpus, model, oper=None):
    figure(figsize=(10, 10), dpi=250)
    print('Corpus processing...')
    vecs = []
    for author in tqdm.tqdm(list(corpus.keys())):
        for book in list(corpus[author].keys()):
            vecs.append(corpus[author][book])
    z_model = model.transform(vecs)
    n = 0
    c18, c19, c20, c21 = [], [], [], []
    for author in tqdm.tqdm(list(corpus.keys())):
        for book in list(corpus[author].keys()):
            if int(book.split('.')[-1]) < 1801:
                c18.append(n)
            if int(book.split('.')[-1]) < 1901 and int(book.split('.')[-1]) >= 1801:
                c19.append(n)
            if int(book.split('.')[-1]) < 2001 and int(book.split('.')[-1]) >= 1901:
                c20.append(n)
            if int(book.split('.')[-1]) >= 2001:
                c21.append(n)
            n+=1
    plt.scatter(z_model[c18, 0], z_model[c18, 1], s=7, color='C0', alpha=0.5, label='XVIII')
    plt.scatter(z_model[c19, 0], z_model[c19, 1], s=7, color='C1', alpha=0.5, label='XIX')
    plt.scatter(z_model[c20, 0], z_model[c20, 1], s=7, color='C2', alpha=0.5, label='XX')
    plt.scatter(z_model[c21, 0], z_model[c21, 1], s=7, color='C3', alpha=0.5, label='XXI')
    plt.title(f'Век написания: {oper} отображение')
    plt.legend(title='Век написания', loc='upper center', bbox_to_anchor=(0.5, -0.05))
    plt.xticks([])
    plt.yticks([])
```

```python
build_model_plot(corpus, mapper, oper='UMAP')
```

```python
from matplotlib.pyplot import figure
def build_books(corpus, target, model, oper=None):
    figure(figsize=(10, 10), dpi=250)
    print('Corpus processing...')
    vecs = []
    for author in tqdm.tqdm(list(corpus.keys())):
        for book in list(corpus[author].keys()):
            if author != target:
                vecs.append(corpus[author][book])
    z_model = model.transform(vecs)
    plt.scatter(z_model[:,0], z_model[:, 1], color='gray', s=7, alpha=0.3)
    print('Done!')
    print('Author books processing...')
    author_vecs = []
    for book in list(corpus[target].keys()):
        author_vecs.append(corpus[target][book])
    z_model = model.transform(author_vecs)
    for n, book in tqdm.tqdm(enumerate(corpus[target])):
        plt.scatter(z_model[n, 0], z_model[n, 1], label=' '.join((book.split('.')[1]).split('_')), s=40, marker=f'${n+1}$')
    print('Done!')
    plt.title(f'{target}: {oper} отображение')
    plt.legend(title='Произведение', loc='upper center', bbox_to_anchor=(0.5, -0.05))
    plt.xticks([])
    plt.yticks([])
```

```python
build_books(corpus, 'Булгаков', mapper, 'UMAP')
```
